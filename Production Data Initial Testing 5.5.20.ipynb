```python
data = data.fillna(0)
data.head(15)
```

```python

data["api_county_code"]= data["api_county_code"].astype(str)
data["api_seq_num"] = data["api_seq_num"].astype(str)
data["sidetrack_num"] = data["sidetrack_num"].astype(str)

m = data["api_county_code"].str.len().max()
n = data['api_seq_num'].str.len().max()
o = 2

data['api_county_code'] = data["api_county_code"].str.rjust(m, '0')
data['api_seq_num'] = data["api_seq_num"].str.rjust(n, '0')
data['sidetrack_num'] = data["sidetrack_num"].str.rjust(o, '0')

data.head()
```
